```python
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from weather.data_generators.national_indicator_fake import generate_fake_national_indicator
```

```python
payload = generate_fake_national_indicator(
    date_start=dt.date(2022, 1, 1),
    date_end=dt.date(2024, 12, 31),
    granularity="month",
    slice_type="day_of_month",
    day_of_month=31
)

df = pd.DataFrame(payload["time_series"])
df["date"] = pd.to_datetime(df["date"])

```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))

# Température observée
plt.plot(df["date"], df["temperature"], label="Température observée", linewidth=2)

# Moyenne baseline
plt.plot(df["date"], df["baseline_mean"], label="Baseline moyenne", linestyle="--")

# Bande écart-type
plt.fill_between(
    df["date"],
    df["baseline_std_dev_lower"],
    df["baseline_std_dev_upper"],
    alpha=0.2,
    label="±1 écart-type",
)

# Min / Max baseline
plt.plot(
    df["date"],
    df["baseline_min"],
    linestyle=":",
    linewidth=1,
    label="Baseline min",
)

plt.plot(
    df["date"],
    df["baseline_max"],
    linestyle=":",
    linewidth=1,
    label="Baseline max",
)

plt.xlabel("Date")
plt.ylabel("Température (°C)")
plt.title("Indicateur thermique national (fake data)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
